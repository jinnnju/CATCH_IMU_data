```python
import pandas as pd

# 1) a_robot에서 필요한 컬럼만 선택
robot_cols = [
    "joint_temperatures_j6",
    "joint_temperatures_j5",
    "joint_temperatures_j4",
    "joint_temperatures_j3",
    "joint_temperatures_j2",
    "joint_temperatures_j1",
    "actual_joint_voltage_j4",
    "actual_joint_voltage_j5",
    "actual_joint_voltage_j2",
    "actual_joint_voltage_j6",
    "actual_joint_voltage_j3",
    "speed_scaling",
    "actual_joint_voltage_j1",
    "actual_robot_voltage",
    "actual_execution_time",
    "actual_tool_accelerometer_z",
    "actual_main_voltage",
    "actual_momentum",
]
a_robot_f = a_robot[robot_cols].copy()

# 2) a_imu에서 첫 번째 컬럼 제외하고 모두 사용
a_imu_ds = a_imu.iloc[:, 1:].copy()

# 3) 길이 맞추기 (짧은 쪽 기준으로 잘라내기)
n = min(len(a_robot_f), len(a_imu_ds))
a_robot_sync = a_robot_f.iloc[:n].reset_index(drop=True)
a_imu_sync = a_imu_ds.iloc[:n].reset_index(drop=True)

# 4) 합치기 (컬럼명 충돌 방지 위해 imu_ 접두사 붙임)
combined = pd.concat([a_robot_sync, a_imu_sync.add_prefix("imu_")], axis=1)

# 최종 결과
print(combined.head())
print("최종 행 개수:", len(combined))
```

catch용 데이터 처리

```python
import numpy as np
import pandas as pd

def IMU_to_CATCH(a_0: pd.DataFrame,
                 a_1: pd.DataFrame,
                 out_path: str,
                 time_col: str = "timestamp") -> pd.DataFrame:
    """
    두 DF(a_0, a_1)의 컬럼명이 동일하다는 전제하에,
    (1) 컬럼명 사전순으로 정렬,
    (2) 각 컬럼별로 a_0 전구간을 먼저, 이어서 a_1 전구간을 쌓고,
    (3) date 인덱스는 a_0→a_1로 '연속'(1..len(a_0)+len(a_1))이 되도록 부여,
    (4) 마지막에 label(문자 그대로) 행 추가: 
        - a_0 구간은 0
        - a_1 구간은 imu_anomaly 열의 값 (0 또는 1)
    형태의 long 포맷(date, data, cols) CSV를 저장한다.
    """
    # 값 컬럼 선택(시간열과 imu_anomaly 제외)
    def value_cols(df: pd.DataFrame) -> list:
        cols = [c for c in df.columns if c not in [time_col, "imu_anomaly"]]
        cols.sort()
        return cols

    cols0 = value_cols(a_0)
    cols1 = value_cols(a_1)
    if set(cols0) != set(cols1):
        raise ValueError(
            "Columns must match (except imu_anomaly).\n"
            f"Missing in a_1: {[c for c in cols0 if c not in cols1]}\n"
            f"Missing in a_0: {[c for c in cols1 if c not in cols0]}"
        )

    cols = cols0  # 정렬된 공통 컬럼 목록
    n0, n1 = len(a_0), len(a_1)
    date0 = np.arange(1, n0 + 1)                 # a_0 구간
    date1 = np.arange(n0 + 1, n0 + n1 + 1)       # a_1 구간

    parts = []
    for col in cols:
        s0 = a_0[col].to_numpy()
        s1 = a_1[col].to_numpy()
        df0 = pd.DataFrame({"date": date0, "data": s0, "cols": col})
        df1 = pd.DataFrame({"date": date1, "data": s1, "cols": col})
        parts.extend([df0, df1])

    # anomaly 라벨 (문자 그대로 'label')
    label0 = pd.DataFrame({"date": date0, "data": 0, "cols": "label"})
    label1 = pd.DataFrame({"date": date1,
                           "data": a_1["imu_anomaly"].to_numpy(),
                           "cols": "label"})

    final_df = pd.concat(parts + [label0, label1], ignore_index=True)
    final_df.to_csv(out_path, index=False, encoding="utf-8-sig")
    return final_df
```
